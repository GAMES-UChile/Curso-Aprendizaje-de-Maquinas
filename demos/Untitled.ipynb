```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
n_times = 1000
time = np.linspace(0,10,n_times)
n_signals = 10
X = np.zeros((n_times,n_signals))
freqs = 2*np.pi*np.random.rand(n_signals)
magn = np.random.rand(n_signals)



for i in range(n_signals):
    X[:,i] = magn[i]*np.cos(freqs[i]*time) 
    
n_channels = 20
mixing_weights = np.random.randn(n_signals, n_channels)
Y = X@mixing_weights
    
plt.figure(figsize=(20,6))
plt.plot(time, X, label = 'series originales');
plt.title('series originales')

plt.figure(figsize=(20,6))
plt.plot(time, Y);
plt.title('series mezcladas')
```

```python
pca = PCA(n_components=10)
principalComponents = pca.fit_transform(Y)
```

```python
plt.figure(figsize=(20,6))
plt.plot(principalComponents);
```
